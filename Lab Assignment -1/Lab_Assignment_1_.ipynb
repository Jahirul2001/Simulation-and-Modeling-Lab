```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task 1: Generate two vectors with 15 random floats each, plot them on the same graph, label the axes, and add a title.


X = np.random.rand(15)
Y = np.random.rand(15)

print("Vector X:\n", X)
print("Vector Y:\n", Y)


plt.plot(X, label='Vector X', color='blue', marker='o')
plt.plot(Y, label='Vector Y', color='green', marker='x')


plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Vectors X and Y")


plt.legend()
plt.grid(True)
plt.show()

```

```python
#Task 2: Create a 4x4 matrix with random values, visualize it as a heatmap, and label the rows and columns.

Matrix = np.random.rand(4, 4)
print("Matrix:\n", Matrix)

plt.figure(figsize=(6, 5))
sns.heatmap(Matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', cbar_kws={'label': 'Value'})

plt.title('Heatmap of 4x4 Matrix', fontsize=14)
plt.xlabel('Column Index', fontsize=12)
plt.ylabel('Row Index', fontsize=12)

plt.show()
```

```python
#Task 3: Generate two 4x4 matrices, perform addition, subtraction, and multiplication, visualize the results using bar plots, and discuss the effects.
Matrix1 = np.random.rand(4, 4)
Matrix2 = np.random.rand(4, 4)

print("Matrix 1:\n", Matrix1)
print("Matrix 2:\n", Matrix2)

Addition = Matrix1 + Matrix2
Subtraction = Matrix1 - Matrix2
Multiplication = Matrix1 * Matrix2

print("Addition:\n", Addition)
print("Subtraction:\n", Subtraction)
print("Multiplication:\n", Multiplication)

add_flat = Addition.flatten()
sub_flat = Subtraction.flatten()
mul_flat = Multiplication.flatten()

indices = np.arange(len(add_flat))

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].bar(indices, add_flat, color='skyblue', edgecolor='black')
axes[0].set_title('Matrix Addition', fontsize=14)
axes[0].set_xlabel('Index', fontsize=12)
axes[0].set_ylabel('Value', fontsize=12)
axes[0].grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

axes[1].bar(indices, sub_flat, color='lightgreen', edgecolor='black')
axes[1].set_title('Matrix Subtraction', fontsize=14)
axes[1].set_xlabel('Index', fontsize=12)
axes[1].set_ylabel('Value', fontsize=12)
axes[1].grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

axes[2].bar(indices, mul_flat, color='salmon', edgecolor='black')
axes[2].set_title('Matrix Multiplication', fontsize=14)
axes[2].set_xlabel('Index', fontsize=12)
axes[2].set_ylabel('Value', fontsize=12)
axes[2].grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

plt.tight_layout()

plt.show()
```
